```python
# Function is for validation generic code.
def validation(prompt = "",cast = None, cond = None, errMsg = None): # accepts four parameter for validating the user input
    # prompt is for string that will be visible to the user
    # cast is the casting of the input value the defalut value is string
    # cond is for condition it checks input with the condition passed by user, if no condition is passed then it chooses default value
    #errMsg is for custom error message  if no message is passed the it will choosed default value.
    while True: # This is infinity loop unless user enters correct value or unless user stops execution
        try: #this is try block for execution and check for error 
            res = (cast or str)(input(prompt)) # If cast is passed then this will check the user input satisfies or else will throw a error. User value are stored in res variable
            assert cond is None or cond(res) # Assert will check for the cond value if none then will return res or else check for the condition passed and if satisfied then return res else give error message
            return res # res has correct value and can be utilized for further computation
        except: # this is catch block if error is generated it will catch it and show the error message
            print(errMsg or "Invalid!!")   # if the error message is passed by user it will show that error message else will show Invalid +

# this function will accept the file name from user
def fileToWork():
    import pandas as pd # import the pandas to read file
    from os.path import exists as file_exists
    path = validation("Enter the File Name",cast=str,cond= lambda x: x.endswith('txt')  and file_exists(x), errMsg= "file not proper") # 'mywords.txt' is example user have to enter 
    # file name if the file is not with txt extension then it will give error.
    d = pd.read_csv(path, header= None,names=["Word", "Meaning"] ) # returns a dataframe which is stored in the variable .the variable is returned when the 
    # function is called
    return d

# this function will choose random word from the file . the parameter is a dataframe with words and file name
def choseWordRandom(file):
    import random # import the random to choose the random word
    from random import shuffle
    wordloc =  random.randrange(len(file)) #choose a random number from 0 to lenght of the file.
    wordChosen = file.iloc[wordloc].Word # use the number to choose the word from the file which is dataframe.
    words = list(wordChosen) # the choosen word is then split in to all character
    shuffle(words) # this shuffle  is used to sramble the words
    displayWord = ''.join(words) # after the scramble join all the character and return sramble word, unscramble word and loc oft the word
    return wordChosen, displayWord, wordloc

file = fileToWork() # this is function to get file for running the function
checkToPlay = input("If you want to start game press 'y'") # Ask if user wnats to play the game
while checkToPlay.lower() == 'y': # check for user response
    wordChosen,displayWord, wordloc = choseWordRandom(file) # word to play the game with
    print("Unscramble the following letters to form a word. Type “?” for the meaning of the unscrambled word:",displayWord)
    attempt = 1 #counter for the attempts
    userAnswer = input("Enter Your answer = ") #accept the user answer
    checkMeaningGiven = 0 # varibale for meaning
    wrongAttemptsExceed = False #if limits exceed
    while userAnswer != wordChosen and attempt < len(wordChosen): # loop to run if word not same 
        if userAnswer == '?': # if ? then show meaning 
            if checkMeaningGiven == 0: # meaning can be asked only once
                print("The word means:", file.iloc[wordloc].Meaning)
                checkMeaningGiven += 1 # increase the counter
                userAnswer = input("Enter Your answer = ")

            else:
                if checkMeaningGiven >= 2: # if meaning asked more than 2 twice then add attempt
                    print(f"Attemp {attempt} out of {len(wordChosen)} added, Wrong!!!")
                    attempt += 1
                    userAnswer = input("Enter Your answer = ")
                else: # give the warning to the user
                    print("You have been given the meaning before. Next time you ask for the meaning it will count as an attempt!")
                    checkMeaningGiven += 1
                    userAnswer = input("Enter Your answer = ")

        else: 
            print(f"Wrong Attemp {attempt} out of {len(wordChosen)}, try again") # if the answer is wrong increase the attempt
            attempt += 1
            userAnswer = input("Enter Your answer = ")

    else: # if the user answer corrects answer then move on
        if userAnswer != wordChosen and attempt == len(wordChosen): # check the attempts not more than length of the words
            print("Wrong, you have exceeded the number of attempts. Bye!")
                  
        elif userAnswer == wordChosen:
            print("You got it!")
        checkToPlay = input("If you want to continue game press 'y'")

else:
    print("GoodBye!!")              

    
```

Question 2

Steps:
1. User have to enter the file name
2. the user will be asked to enter the line he wants to compare
3. User will the cosine similarity for all the poems in the file

Algorithm:
1. enter file and check the file extension and if the file exists. the user will be asked again to enter the file name if not correct
2. Once file name is checked then file is read using pandas and convert into dataframe
3. Create dictionary for the words in the file uisng \n and new dicrectory also store occurrence count as value in dictionary
4. iterate the dictioanry and store the values as vector array and also store the dictionary into the array
5. then add the dictionary array and vector of array to the dataframe.
6. Now use this values of vector that are in the dataframe to calculate the cosine similarity
7. If the lenght of the vector array is not same the append the values to create array of same length
8. return the cosine similarity for all the poems

```python
# Function is for validation generic code.
def validation(prompt = "",cast = None, cond = None, errMsg = None): # accepts four parameter for validating the user input
    # prompt is for string that will be visible to the user
    # cast is the casting of the input value the defalut value is string
    # cond is for condition it checks input with the condition passed by user, if no condition is passed then it chooses default value
    #errMsg is for custom error message  if no message is passed the it will choosed default value.
    while True: # This is infinity loop unless user enters correct value or unless user stops execution
        try: #this is try block for execution and check for error 
            res = (cast or str)(input(prompt)) # If cast is passed then this will check the user input satisfies or else will throw a error. User value are stored in res variable
            assert cond is None or cond(res) # Assert will check for the cond value if none then will return res or else check for the condition passed and if satisfied then return res else give error message
            return res # res has correct value and can be utilized for further computation
        except: # this is catch block if error is generated it will catch it and show the error message
            print(errMsg or "Invalid!!")   # if the error message is passed by user it will show that error message else will show Invalid +

# this function will accept file and read and convert in to the dataframe with Author and lines as the column name. the separator is the ':'
def fileToWork():
    import pandas as pd
    from os.path import exists as file_exists # check if the file exist
    path = validation("Enter the File Name",cast=str,cond= lambda x: x.endswith('txt')  and file_exists(x), errMsg= "file not proper") # '"poetry_lines.txt" is example user have to enter '
    data = pd.read_csv(path, sep=":",header = None, names=["Author", "lines"]) # convert into the dataframe with author and lines as the column name and return the data
    return data

# this function will create the dictionary
def createDict(lines):
    wordDict = {} # intialization
    word = lines.strip().split() # split the sentence and store to the list 
    for i in word: #iterate over the list of words collected from the list
        if i in wordDict: # convert list to dictionary and word as the key and occurrence count 
            wordDict[i] += 1
        else: # if occurr first time then add to the dictionary with 1 as occurrence    
            wordDict[i]  = 1
    return wordDict            # return the dictionary

# function will collect dictionary and the store the values of dictionary
def generateVector(data):
    # intialize the array to variable
    dictArray = [] 
    authors = []   
    vectorArray = [] 
    for i in range(len(data)): # iterate over the line given
        authors.append(data.iloc[i].Author.strip()) # append the author name to the list
        wordDict = createDict(data.iloc[i].lines) # function call for the getting the word dictionary
        vectorArray.append(list(wordDict.values())) # this will store the values for all the poems in the file
        dictArray.append(wordDict) # add dictionary to the array
    data["Dict"] = list(dictArray) # add the array to the data frame
    data["Vector"] = list(vectorArray)# add the array to the dataframe
# return the data frame
    return data

def userData(data):
    # import required libraries
    import numpy as np
    from numpy.linalg import norm
    simliar = 0 #intialize the counter
    userInput = input("Enter the sentence") #take input from the user
    userDict = createDict(userInput) #create dictionary of words and their occurrence for the user input line
    userVector = list(userDict.values()) # create the vector from the user entered dictionary
    for i in range(len(data)):
        ml = max(len(data.iloc[i].Vector),len(userVector)) # check for the max length from user line vector or file line vector
        A = np.concatenate((data.iloc[i].Vector , np.zeros(ml-len(data.iloc[i].Vector)))) # add zeros for the place that doesnt match
        B = np.concatenate((userVector , np.zeros(ml-len(userVector))))# add zeros
        # compute cosine similarity
        cosine = np.dot(A,B)/(norm(A)*norm(B)) # formula for the cosine similarity
        print(f"{data.iloc[i].Author}:{cosine:.2f}") # show cosine similarity for all the poets
        if simliar < cosine:
            simliar = cosine
            closest = i
    print("The closest poem is:\n", data.iloc[closest].Author,":",data.iloc[closest].lines)    

def main():
    data  = fileToWork()
    vectorData = generateVector(data)
    userData(vectorData)
main()
```

Question 3

Steps:
1. The user has to install all the required library
2. the libraries used are sqlite3, bs4, requests

Algorithm:
1. we are using three different links for three different commodities. user can change the link as needed.
2. we are requesting the page from the link
3. Parse the response of the user to the html
4. we can iterate the html like tree form and use the tags to reach the location .
5. Using class as uniques identifier for getting the value from the page.(All page has same html view so using core commodity page)
6. store all the values into the array name and prices
7. then open a file in write mode and write all the values stored in the array name and prices.
8. Open a connecntion with the sqllite and check if the table exists if yes drop the table and create new table.
9. Insert the data that are in the list names and prices
10. close the connection

```python
# !pip install bs4

# check for the file and replace check is not tested
def getData():
    # import the file 
    import requests
    from bs4 import BeautifulSoup
    #initalize to array 
    name = []
    prices = []
    # link for the crude oil to get the data
    urlCrud = 'https://finance.yahoo.com/quote/CL%3DF'
    # tag to get the prices from the page
    tagValue = 'div'
    tagName = 'h1'
    # using class as the identifier to get the value from specific section
    findValue = {"class":"D(ib) Mend(20px)"}
    findName = {"class":"D(ib) Fz(18px)"}
    # request the page using the url
    respOil = requests.get(url= urlCrud)
    # this will parse in the html file
    beautifyOil = BeautifulSoup(respOil.text,'html.parser')
    # find the tag with class to get the value as name
    nameC = beautifyOil.find(tagName,findName).text
    # use find the to go to the section which has value iterate over all and observed the first one will return price
    priceOil = beautifyOil.find(tagValue,findValue).find_all('fin-streamer')[0].text
    # append to the array
    name.append(nameC)
    prices.append(priceOil)

    # link for the Gold to get the data
    urlGold = 'https://finance.yahoo.com/quote/GC=F'
    # request the page using the url
    respGold = requests.get(url= urlGold)
    # this will parse in the html file
    beautifyGold = BeautifulSoup(respGold.text,'html.parser')
    # find the tag with class to get the value as name
    nameG = beautifyGold.find(tagName,findName).text
    # use find the to go to the section which has value iterate over all and observed the first one will return price
    priceGold = beautifyGold.find(tagValue,findValue).find_all('fin-streamer')[0].text
    # append to the array
    name.append(nameG)
    prices.append(priceGold)

    # link for the Gold to get the data
    urlSilver = 'https://finance.yahoo.com/quote/SI=F'
    # request the page using the url
    respSilver = requests.get(url= urlSilver)
    # this will parse in the html file
    beautifysilver = BeautifulSoup(respSilver.text,'html.parser')
    # find the tag with class to get the value as name
    nameS = beautifysilver.find(tagName,findName).text
    # use find the to go to the section which has value iterate over all and observed the first one will return price
    priceSilver = beautifysilver.find(tagValue,findValue).find_all('fin-streamer')[0].text
    # append to the array
    name.append(nameS)
    prices.append(priceSilver)
    return name,prices

#function parameter is name and price array
def generateFile(name,prices):
    f = open('commodity_prices.txt', 'w+') # open the file with write mode
    f.write("Commodity\t:" + "\tPrices\n") # write column name in the array
    for i in range(len(name)): #iterate over name and prices array
        f.write( name[i] +"\t:\t" + prices[i] +"\n")
    f.close()    # close once the data is written

# function will create connection with the sqlite3
def createConnection():
    import sqlite3
# specify the schema name
    conn = sqlite3.connect('CommodityDatabase.sqlite')
    # create a cursor 
    cur = conn.cursor()
    # check if the table exists if yes then drop
    cur.execute("DROP TABLE IF EXISTS CommodityTable")
    # create a new table to store data wint ticker and price as name of columns
    cur.execute('CREATE TABLE CommodityTable (Ticker TEXT, Price REAL)')
    # commit the change and return the connection object
    conn.commit()
    return  conn


def insertValue(conn,val):    
    # using the connection object and values insert into the table
    sql = ''' INSERT INTO CommodityTable(Ticker,Price)VALUES(?,?) '''
    # create cursor instance
    cur = conn.cursor()
    #execute the query
    cur.execute(sql, val)
    # commit the changes
    conn.commit()

def main():
    name,prices = getData()
    generateFile(name,prices)
    con = createConnection()
    for i in range(len(name)):
        val = (name[i],prices[i])
        insertValue(con,val)
    con.close()    
main()
    
```
